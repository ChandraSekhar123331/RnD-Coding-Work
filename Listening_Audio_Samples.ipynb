```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import torchaudio
import librosa
import IPython.display as ipd
import numpy as np
import json
```

```python
audio_path = '/home/mayank/MTP/begin_again/Error-Driven-ASR-Personalization/mozilla/cv-corpus-7.0-2021-07-21/en/wav/common_voice_en_619470.wav'
# audio_path = audio_path.replace('/wav/', '/clips/').replace('.wav', '.mp3')
# audio_path = audio_path.replace('/clips/', '/wav/').replace('.mp3', '.wav')
print(audio_path)
```

```python
speech, sr = torchaudio.load(audio_path)
print(f"sampling rate is = {sr}")
speech = speech[0].numpy().squeeze()
speech = librosa.resample(np.asarray(speech), sr, 16_000)
ipd.Audio(data = np.asarray(speech), autoplay = True, rate = 16_000)
```
